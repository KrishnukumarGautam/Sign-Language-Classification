```python
# ==============================
# ğŸ“Œ SIGN LANGUAGE RECOGNITION PROJECT (Top-3 + Misclassification Check)
# ==============================

# 1ï¸âƒ£ Install dependencies
!pip install tensorflow opencv-python matplotlib

# 2ï¸âƒ£ Upload & Extract Dataset
from google.colab import files
import zipfile

print("ğŸ“‚ Please upload your 'sign-language-mnist.zip' file (from Kaggle)")
uploaded = files.upload()

for fn in uploaded.keys():
    with zipfile.ZipFile(fn, 'r') as zip_ref:
        zip_ref.extractall("sign_mnist_data")

print("âœ… Dataset extracted to 'sign_mnist_data/' folder")

# 3ï¸âƒ£ Load Dataset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Load CSV files from extracted dataset
train_df = pd.read_csv("sign_mnist_data/sign_mnist_train.csv")
test_df  = pd.read_csv("sign_mnist_data/sign_mnist_test.csv")

# Features and labels
X_train = train_df.iloc[:,1:].values.reshape(-1,28,28,1) / 255.0
y_train = to_categorical(train_df['label'].values)

X_test  = test_df.iloc[:,1:].values.reshape(-1,28,28,1) / 255.0
y_test  = to_categorical(test_df['label'].values)

print("âœ… Training set:", X_train.shape, y_train.shape)
print("âœ… Test set:", X_test.shape, y_test.shape)

# 4ï¸âƒ£ Build CNN Model
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(25, activation='softmax')   # 25 sign language classes
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# 5ï¸âƒ£ Train Model
history = model.fit(X_train, y_train,
                    epochs=15, batch_size=128,
                    validation_data=(X_test, y_test))

# Evaluate final accuracy
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"ğŸ¯ Final Test Accuracy: {test_acc*100:.2f}%")

# 6ï¸âƒ£ Plot Accuracy and Loss
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Accuracy
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend()
plt.title("Model Accuracy")

# Loss
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.title("Model Loss")

plt.show()

# 7ï¸âƒ£ Upload & Predict Single Custom Image (Top-3 + Misclassification)
import cv2

# Mapping label indexes to letters (skip J & Z because not in dataset)
labels_map = {i: chr(65+i) for i in range(25)}

print("ğŸ“¤ Upload a hand sign image for prediction")
uploaded_img = files.upload()

for fn in uploaded_img.keys():
    img = cv2.imread(fn, cv2.IMREAD_GRAYSCALE)
    img_resized = cv2.resize(img, (28,28)) / 255.0
    img_input = img_resized.reshape(1,28,28,1)

    # Prediction
    pred = model.predict(img_input)[0]

    # Top 3 predictions
    top_indices = pred.argsort()[-3:][::-1]
    top_probs = pred[top_indices]

    print("\nğŸ” Top 3 Predictions:")
    for i, idx in enumerate(top_indices):
        print(f"{i+1}) {labels_map[idx]} - {top_probs[i]*100:.2f}%")

    # True label input
    true_label = input("\nâœï¸ Enter the true label (A-Y, except J & Z): ").upper()

    top_predicted = labels_map[top_indices[0]]
    if top_predicted == true_label:
        print(f"\nâœ… Correct! Model predicted: {top_predicted}")
    else:
        print(f"\nâŒ Wrong Prediction! Model predicted: {top_predicted}, but true label was: {true_label}")

    # Show uploaded image
    plt.imshow(img, cmap='gray')
    plt.axis("off")
    plt.title(f"Predicted: {top_predicted} | True: {true_label}")
    plt.show()
```
